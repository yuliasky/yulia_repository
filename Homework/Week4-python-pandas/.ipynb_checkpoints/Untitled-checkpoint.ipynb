```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
school_data_complete.count()
```

```python
nb_schools=len(school_data_complete["school_name"].unique())
nb_schools
```

```python
nb_students=len(school_data_complete["student_name"])
nb_students
```

```python
school_data_complete.dtypes
```

```python
total_budget=school_data["budget"].sum()
total_budget
```

```python
av_math_score=school_data_complete["math_score"].mean()
av_math_score
```

```python
av_read_score=school_data_complete["reading_score"].mean()
av_read_score
```

```python
av_pass_rate=(av_math_score + av_read_score)/2
av_pass_rate
```

```python
pass_math=(len(school_data_complete.loc[school_data_complete["math_score"]>=70,:]))/nb_students*100
pass_math
```

```python
pass_read=(len(school_data_complete.loc[school_data_complete["reading_score"]>=70,:]))/nb_students*100
pass_read
```

```python
df_dist_sum=pd.DataFrame({"Total Schools":[nb_schools],"Total Students":[nb_students],"Total Budget":[total_budget],
                          "Average Math Score":[av_math_score],"Average Reading Score":[av_read_score],
                          "% Passing Math":[pass_math],"% Passing Reading":[pass_read],
                          "% Overall Passing Rate":[av_pass_rate]})
df_dist_sum
```

```python
school_group=school_data_complete.groupby("school_name")
school_group.count().head()
```

```python
school_group_stu=pd.DataFrame(school_group["student_name"].count())
school_group_stu.head(15)
```

```python
school_group_df=school_group_stu.rename(columns={"student_name":"Total Students"})
school_group_df
```

```python
school_group_df=school_group_df.merge(school_data, on="school_name")
school_group_df.head(15)
```

```python
school_group_df=school_group_df.set_index("school_name")
school_group_df
```

```python
school_group_df=school_group_df.loc[:,["type","Total Students","budget"]]
school_group_df
```

```python
school_group_df["Per Student Budget"]=school_group_df["budget"]/school_group_df["Total Students"]
school_group_df["Per Student Budget"]=school_group_df["Per Student Budget"].map('${:.2f}'.format)
school_group_df
```

```python
school_group_av_math=pd.DataFrame(school_group["math_score"].mean())
school_group_av_math.head(15)
```

```python
school_group_df=school_group_df.merge(school_group_av_math, on="school_name")
school_group_df.head(15)
```

```python
school_group_av_read=pd.DataFrame(school_group["reading_score"].mean())
school_group_av_read.head(15)
```

```python
school_group_df=school_group_df.merge(school_group_av_read, on="school_name")
school_group_df.head(15)
```

```python
school_pass_math=school_data_complete.loc[school_data_complete["math_score"]>=70,:]
school_pass_math.head()
```

```python
school_pass_math_group=school_pass_math.groupby("school_name")
school_pass_math_group.count().head()
```

```python
school_group_pass_math=pd.DataFrame(school_pass_math_group["math_score"].count())
school_group_pass_math.head(15)
```

```python
school_group_pass_math=school_group_pass_math.rename(columns={"math_score":"count_math_score_pass"})
school_group_pass_math
```

```python
school_group_df["% Passing Math"]=(school_group_pass_math["count_math_score_pass"]/school_group_df["Total Students"])*100
school_group_df
```

```python
school_pass_read=school_data_complete.loc[school_data_complete["reading_score"]>=70,:]
school_pass_read.head()
```

```python
school_pass_read_group=school_pass_read.groupby("school_name")
school_pass_read_group.count().head()
```

```python
school_group_pass_read=pd.DataFrame(school_pass_read_group["reading_score"].count())
school_group_pass_read.head(15)
```

```python
school_group_pass_read=school_group_pass_read.rename(columns={"reading_score":"count_reading_score_pass"})
school_group_pass_read
```

```python
school_group_df["% Passing Reading"]=(school_group_pass_read["count_reading_score_pass"]/school_group_df["Total Students"])*100
school_group_df
```
