```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
# just to check the count
school_data_complete.count()
```

```python
# Get nb of schools
nb_schools=len(school_data_complete["school_name"].unique())
nb_schools
```

```python
# Get nb of students
nb_students=len(school_data_complete["student_name"])
nb_students
```

```python
# check types
school_data_complete.dtypes
```

```python
# Calculate Total Busget
total_budget=school_data["budget"].sum()
total_budget
```

```python
# Calculate Avergate Math Score
av_math_score=school_data_complete["math_score"].mean()
av_math_score
```

```python
# Calculate Average Reading Score
av_read_score=school_data_complete["reading_score"].mean()
av_read_score
```

```python
# Calculate Overall Passing rate
av_pass_rate=(av_math_score + av_read_score)/2
av_pass_rate
```

```python
# Caculate % students with passing math score >70
pass_math=(len(school_data_complete.loc[school_data_complete["math_score"]>=70,:]))/nb_students*100
pass_math
```

```python
# Calculate % students with passing reading score >70
pass_read=(len(school_data_complete.loc[school_data_complete["reading_score"]>=70,:]))/nb_students*100
pass_read
```

```python
# Create DF with above results
df_dist_sum=pd.DataFrame({"Total Schools":[nb_schools],"Total Students":[nb_students],"Total Budget":[total_budget],
                          "Average Math Score":[av_math_score],"Average Reading Score":[av_read_score],
                          "% Passing Math":[pass_math],"% Passing Reading":[pass_read],
                          "% Overall Passing Rate":[av_pass_rate]})
df_dist_sum
```
